```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from string import punctuation
import re

# Ensure you've downloaded the necessary NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

# Specify the path to your uncleaned text file
text_file_path = '/content/drive/MyDrive/Desktop/New_youtube_fafsa_comments1.txt'

# Read the text file into a list
with open(text_file_path, 'r', encoding='utf-8') as file:
    comments = file.readlines()

# Define the cleaning functions (lowercasing, removing stop words, punctuation, white spaces, lemmatization)
def clean_comment(comment):
    # Lowercase the comment
    comment = comment.lower()

    # Remove punctuation
    comment = re.sub(r'[^\w\s]', '', comment)

    # Remove white spaces and strip
    comment = re.sub(r'\s+', ' ', comment).strip()

    # Tokenize the comment
    tokens = word_tokenize(comment)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # Join the tokens back into a cleaned comment
    cleaned_comment = ' '.join(tokens)

    return cleaned_comment

# Clean the comments
cleaned_comments = [clean_comment(comment) for comment in comments]

# Convert the cleaned comments to a DataFrame
comments_df_cleaned = pd.DataFrame({'Comment': cleaned_comments})

# Specify the path to save the cleaned comments as a CSV file
cleaned_csv_file_path = '/content/drive/MyDrive/Desktop/cleaned_youtube_fafsa_comments1.csv'

# Save the cleaned DataFrame to a CSV file
comments_df_cleaned.to_csv(cleaned_csv_file_path, index=False)

print(f"Cleaned comments saved to CSV at: {cleaned_csv_file_path}")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
import pandas as pd

# Define a function to clean and lemmatize a single comment
def clean_and_lemmatize_comment(comment):
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()
    # Remove punctuation and lowercase words
    comment = comment.translate(str.maketrans('', '', string.punctuation)).lower()
    # Tokenize and remove stop words and lemmatize
    tokens = word_tokenize(comment)
    cleaned_tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalpha() and word not in stop_words]
    # Rejoin into a string
    cleaned_comment = ' '.join(cleaned_tokens)
    return cleaned_comment

# Function to process all comments from a file
def process_comments(file_path):
    with open(file_path, 'r') as file:
        comments = file.readlines()  # Read the comments into a list

    # Clean and lemmatize each comment
    cleaned_comments = [clean_and_lemmatize_comment(comment) for comment in comments]
    return cleaned_comments

# Path to the text file
file_path = '/content/drive/MyDrive/Desktop/New_youtube_fafsa_comments1.txt'
cleaned_comments = process_comments(file_path)

# Convert to DataFrame and save to CSV
df = pd.DataFrame({'cleaned_comments': cleaned_comments})
output_path = '/content/drive/MyDrive/Desktop/cleaned_new_youtube_fafsa_comments1.csv'
df.to_csv(output_path, index=False)
```

```python
import string
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import download

# Download necessary NLTK resources
download('punkt')
download('stopwords')
download('wordnet')

# Function to clean comments
def clean_comment(comment):
    # Convert to lower case
    lower = comment.lower()
    # Remove punctuation
    no_punctuation = lower.translate(str.maketrans('', '', string.punctuation))
    # Tokenization
    tokens = word_tokenize(no_punctuation)
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(word) for word in tokens]
    # Rejoin words
    return ' '.join(lemmatized)

# Path to the comments file
comments_file_path = '/content/drive/MyDrive/Desktop/H_iS7gmQd90/H_iS7gmQd90_comments.txt'

# Load the comments from the file
with open(comments_file_path, 'r') as file:
    comments = file.readlines()

# Strip whitespace and clean the comments
cleaned_comments = [clean_comment(comment.strip()) for comment in comments]

# Convert to DataFrame
comments_df = pd.DataFrame(cleaned_comments, columns=['Cleaned_Comment'])

# Define the path for the output CSV file
cleaned_comments_csv_path = '/content/drive/MyDrive/Desktop/H_iS7gmQd90/H_iS7gmQd90_cleaned_comments.csv'

# Save the cleaned comments to a CSV file
comments_df.to_csv(cleaned_comments_csv_path, index=False)

print(f'Cleaned comments saved to: {cleaned_comments_csv_path}')
```

```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Ensure you have the necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Function to clean and lemmatize comments
def clean_and_lemmatize(comment):
    # Convert to lowercase
    comment = comment.lower()
    # Remove punctuation
    comment = comment.translate(str.maketrans('', '', string.punctuation))
    # Tokenize
    tokens = word_tokenize(comment)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(w) for w in tokens]
    return ' '.join(lemmatized)

# Path to the comments text file
comments_file_path = '/content/drive/MyDrive/Desktop/Pn4OECMTh5w/comments.txt'

# Load the comments from the file
with open(comments_file_path, 'r') as file:
    comments = file.readlines()

# Clean and lemmatize the comments
cleaned_comments = [clean_and_lemmatize(comment) for comment in comments]

# Save the cleaned comments to a CSV file
cleaned_comments_csv_path = '/content/drive/MyDrive/Desktop/Pn4OECMTh5w/cleaned_comments.csv'
pd.DataFrame(cleaned_comments, columns=['Cleaned_Comment']).to_csv(cleaned_comments_csv_path, index=False)

print(f'Cleaned comments saved to {cleaned_comments_csv_path}')
```
